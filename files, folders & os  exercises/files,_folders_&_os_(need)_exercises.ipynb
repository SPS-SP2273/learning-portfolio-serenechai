```python
import os
import shutil
import glob

cities = ['Jakarta', 'Karachi', 'Manila', 'Melbourne', 'Osaka', 'Shanghai', 'Singapore', 'Tokyo']

for city in cities:
    city_files = glob.glob(f'*-{city.lower()}*')
    for file in city_files:
        current_path_of_files = os.path.basename(file)
        new_path_of_files = os.path.join(city, current_path_of_files)
        shutil.copy(file, new_path_of_files)
```
